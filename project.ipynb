```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the scv dataframe
heart_data = pd.read_csv('dataset.csv')
```

```python
heart_data.head()
```

```python
#print last 5 rows
heart_data.tail()
```

```python
#number of rows and columns in a given daatset
heart_data.shape
```

```python
#getting some infromation from data
heart_data.info()
```

```python
#checking for missing values 
heart_data.isnull().sum()
```

```python
#stastical measures of the data
heart_data.describe()
```

```python
#the distribution of target variables
heart_data['target'].value_counts()
```

```python
#splitting the features and target
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
#spllitting the data into traning data and test data
X_train,X_test,Y_train,Y_test=train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
#Logistic regression 
model=LogisticRegression()
```

```python
#we have to train the model with traning data(finds the relationship between features of the model and target of the model)
model.fit(X_train, Y_train)
```

```python
#Evalating the model
#accuracy on traning data
X_train_prediction=model.predict(X_train)
traning_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print(' Accuracy on traning data : ' ,traning_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print(' Accuracy on testing data : ' ,test_data_accuracy)
```

```python
#Building the predictive System
#to get input features
input_data=(37,1,4,140,207,0,0,130,1,1.5,2)

#change the input data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for only 1 data point
input_data_reshaped=input_data_as_numpy_array.reshape(1, -1)

prediction= model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
    print('The Person without heart disease')
else:
    print('The Person with heart disease')
```
